# CSV Data Exploration
Analysis of `photos_metadatas_filtered.csv`

## Load & inspect data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

```python
df = pd.read_csv('data/photos_metadatas_filtered.csv')

print("Dataset Shape:", df.shape)
print("\nFirst few rows:")
df.head()
```

```python
# Column information
print("Column Names and Types:")
print(df.dtypes)
print("\n" + "="*50)
print("Column Info:")
df.info()
```

```python
# Statistical summary
print("Statistical Summary:")
df.describe(include='all')
```

```python
# Unique values for categorical columns
print("Unique Values per Column:")
for col in df.columns.sort_values():
    unique_count = df[col].nunique()
    print(f"{col}: {unique_count} unique values")
```

```python
# Missing values analysis
print("Missing Values:")
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Column': missing.index, 'Missing_Count': missing.values, 'Missing_Percentage': missing_pct.values})
print(missing_df[missing_df['Missing_Count'] > 0] if missing_df['Missing_Count'].sum() > 0 else "No missing values!")
```

```python
# Column completeness summary
completeness = (1 - df.isnull().sum() / len(df)) * 100
high_completeness = completeness[completeness >= 90].sort_values(ascending=False)
low_completeness = completeness[completeness < 90].sort_values(ascending=True)

print(f"Columns with >= 90% completion: {len(high_completeness)}")
print(f"Columns with < 90% completion: {len(low_completeness)}")

print(f"\nColumns with < 50% completion:")
print(low_completeness[low_completeness < 50].sort_values())

print(f"\nCore fields completeness:")
core_fields = ['DateTimeOriginal', 'ISO', 'Aperture', 'FNumber', 'ShutterSpeed', 'Model', 'LensID', 'Lens', 'GPS', 'Keywords', 'Rating']
for field in core_fields:
    if field in df.columns:
        comp = completeness[field]
        print(f"  {field}: {comp:.1f}%")
```

```python
print("\n" + "=" * 70)
print("COMPLETENESS CATEGORIES")
print("=" * 70)

completeness = (1 - df.isnull().sum() / len(df)) * 100

# Break into categories
perfect = completeness[completeness == 100.0]
nearly_perfect = completeness[(completeness >= 99) & (completeness < 100)]
excellent = completeness[(completeness >= 95) & (completeness < 99)]
good = completeness[(completeness >= 90) & (completeness < 95)]
partial = completeness[(completeness >= 50) & (completeness < 90)]
sparse = completeness[completeness < 50]

print(f"\n🟢 PERFECT (100%): {len(perfect)} columns")
print(f"   These have every row filled with data")

print(f"\n💚 NEARLY PERFECT (99-99.9%): {len(nearly_perfect)} columns")
print(f"   Missing <60 records total")
print(f"   Example: {nearly_perfect.index[:3].tolist() if len(nearly_perfect) > 0 else 'None'}")

print(f"\n💙 EXCELLENT (95-98.9%): {len(excellent)} columns")
print(f"   Missing <3,000 records (96% of data or more)")
print(f"   Example: {excellent.index[:3].tolist() if len(excellent) > 0 else 'None'}")

print(f"\n🟡 GOOD (90-94.9%): {len(good)} columns")
print(f"   Missing 5-10% but still usable")
print(f"   Example: {good.index[:3].tolist() if len(good) > 0 else 'None'}")

print(f"\n🟠 PARTIAL (50-89.9%): {len(partial)} columns")
print(f"   Should only use for exploratory lookup")
print(f"   Example: {partial.index[:5].tolist() if len(partial) > 0 else 'None'}")

print(f"\n🔴 SPARSE (<50%): {len(sparse)} columns")
print(f"   Not recommended for main analysis")
print(f"   Example: {sparse.index[:5].tolist() if len(sparse) > 0 else 'None'}")

print(f"\n{'='*70}")
print(f"✅ CONCLUSION: {len(perfect) + len(nearly_perfect) + len(excellent)} columns have 95%+ data")
print(f"✅ Your core EXIF columns (dates, ISO, aperture, camera) are all excellent")
```

```python
fig, ax = plt.subplots(figsize=(14, 28))

completeness = (1 - df.isnull().sum() / len(df)) * 100
completeness_sorted = completeness.sort_values(ascending=True)

# Assign colors based on completeness categories with different green shades
def get_color(value):
    if value == 100.0:
        return '#006400'  # Dark green (Perfect: 100%)
    elif value >= 99:
        return '#228B22'  # Forest green (Nearly Perfect: 99-99.9%)
    elif value >= 95:
        return '#3CB371'  # Medium sea green (Excellent: 95-98.9%)
    elif value >= 90:
        return '#90EE90'  # Light green (Good: 90-94.9%)
    elif value >= 50:
        return '#FFB347'  # Light orange (Partial: 50-89.9%)
    else:
        return '#DC143C'  # Crimson red (Sparse: <50%)

colors = [get_color(x) for x in completeness_sorted.values]

ax.barh(range(len(completeness_sorted)), completeness_sorted.values, color=colors, edgecolor='black', alpha=0.8, height=0.88)
ax.set_yticks(range(len(completeness_sorted)))
ax.set_yticklabels(completeness_sorted.index, fontsize=5.5)
ax.set_xlabel('Completeness (%)', fontsize=12, fontweight='bold')
ax.set_title('Column Completeness Summary: Metadata Quality Assessment', fontsize=14, fontweight='bold', pad=20)

# Add threshold lines with distinct visual properties
line_100 = ax.axvline(100, color='darkgreen', linestyle='-', linewidth=2.5, alpha=0.8, label='Perfect (100%)')
line_99 = ax.axvline(99, color='green', linestyle=':', linewidth=2, alpha=0.7, label='Nearly Perfect (99%)')
line_95 = ax.axvline(95, color='limegreen', linestyle=':', linewidth=1.5, alpha=0.6, label='Excellent (95%)')
line_90 = ax.axvline(90, color='blue', linestyle='--', linewidth=2.5, alpha=0.8, label='Good Threshold (90%)')
line_50 = ax.axvline(50, color='red', linestyle='--', linewidth=2, alpha=0.7, label='Sparse Threshold (50%)')

ax.set_xlim(0, 105)
ax.grid(axis='x', alpha=0.3, linestyle='-', linewidth=0.5)

# Create legend for bar colors (completeness categories)
color_legend_elements = [
    Patch(facecolor='#006400', edgecolor='black', linewidth=1.5, label='Perfect (100%)'),
    Patch(facecolor='#228B22', edgecolor='black', linewidth=1.5, label='Nearly Perfect (99-99.9%)'),
    Patch(facecolor='#3CB371', edgecolor='black', linewidth=1.5, label='Excellent (95-98.9%)'),
    Patch(facecolor='#90EE90', edgecolor='black', linewidth=1.5, label='Good (90-94.9%)'),
    Patch(facecolor='#FFB347', edgecolor='black', linewidth=1.5, label='Partial (50-89.9%)'),
    Patch(facecolor='#DC143C', edgecolor='black', linewidth=1.5, label='Sparse (<50%)'),
]

# Create legend for threshold lines
line_legend_elements = [
    Line2D([0], [0], color='darkgreen', linewidth=2.5, linestyle='-', label='100% Threshold'),
    Line2D([0], [0], color='green', linewidth=2, linestyle=':', label='99% Threshold'),
    Line2D([0], [0], color='limegreen', linewidth=1.5, linestyle=':', label='95% Threshold'),
    Line2D([0], [0], color='blue', linewidth=2.5, linestyle='--', label='90% Threshold'),
    Line2D([0], [0], color='red', linewidth=2, linestyle='--', label='50% Threshold'),
]

# Add both legends
legend1 = ax.legend(handles=color_legend_elements, loc='lower center', fontsize=9, 
                    title='Data Completeness Levels', title_fontsize=10, framealpha=0.95, ncol=2)
ax.add_artist(legend1)

legend2 = ax.legend(handles=line_legend_elements, loc='lower right', fontsize=9, 
                    title='Quality Thresholds', title_fontsize=10, framealpha=0.95)

plt.tight_layout()
plt.savefig('plots/column_completeness.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: plots/column_completeness.png")
```

## Temporal Analysis

```python
# Parse dates and extract temporal info using DateTimeOriginal
df['DateTime'] = pd.to_datetime(df['DateTimeOriginal'], format='%Y:%m:%d %H:%M:%S', errors='coerce')
df['Year_parsed'] = df['DateTime'].dt.year
df['Month_parsed'] = df['DateTime'].dt.month
df['Hour_parsed'] = df['DateTime'].dt.hour

print("Temporal Statistics:")
print(f"Start Year: {int(df['Year_parsed'].min())}")
print(f"End Year: {int(df['Year_parsed'].max())}")
print(f"\nPhotos per Year:")
print(df['Year_parsed'].value_counts().sort_index())
print(f"\nPhotos per Month (overall):")
print(df['Month_parsed'].value_counts().sort_index())
print(f"\nPhotos per Hour of Day:")
hourly = df['Hour_parsed'].value_counts().sort_index()
print(hourly)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Photos per year
yearly = df['Year_parsed'].value_counts().sort_index()
axes[0, 0].bar(yearly.index, yearly.values, color='steelblue', edgecolor='black', alpha=0.7)
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Photo Count')
axes[0, 0].set_title('Photos per Year')
axes[0, 0].grid(axis='y', alpha=0.3)

# Photos per month
monthly = df['Month_parsed'].value_counts().sort_index()
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
axes[0, 1].bar(range(1, 13), [monthly.get(i, 0) for i in range(1, 13)], color='coral', edgecolor='black', alpha=0.7)
axes[0, 1].set_xlabel('Month')
axes[0, 1].set_ylabel('Photo Count')
axes[0, 1].set_title('Photos per Month (Aggregated)')
axes[0, 1].set_xticks(range(1, 13))
axes[0, 1].set_xticklabels(month_labels, rotation=45)
axes[0, 1].grid(axis='y', alpha=0.3)

# Photos per hour
hourly = df['Hour_parsed'].value_counts().sort_index()
axes[1, 0].plot(hourly.index, hourly.values, marker='o', linewidth=2, markersize=4, color='darkgreen')
axes[1, 0].fill_between(hourly.index, hourly.values, alpha=0.3, color='green')
axes[1, 0].set_xlabel('Hour of Day')
axes[1, 0].set_ylabel('Photo Count')
axes[1, 0].set_title('Bimodal Distribution: Photos per Hour of Day')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].set_xticks(range(0, 24, 2))

# Year-month heatmap style
year_month = df.groupby(['Year_parsed', 'Month_parsed']).size().unstack(fill_value=0)
im = axes[1, 1].imshow(year_month.T, aspect='auto', cmap='YlOrRd', interpolation='nearest')
axes[1, 1].set_xlabel('Year')
axes[1, 1].set_ylabel('Month')
axes[1, 1].set_title('Photo Distribution Heatmap (Year × Month)')
axes[1, 1].set_xticks(range(len(year_month.index)))
axes[1, 1].set_xticklabels([int(y) for y in year_month.index])
axes[1, 1].set_yticks(range(12))
axes[1, 1].set_yticklabels(month_labels)
plt.colorbar(im, ax=axes[1, 1])

plt.tight_layout()
plt.savefig('plots/temporal_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: plots/temporal_distribution.png")
```

## Parquet Data Overview

```python
# Read the specific parquet file with histogram RGB data
parquet_path = './data/photos_metadatas_with_hist_rgb128.parquet'
df_parquet = pd.read_parquet(parquet_path)

print("=" * 80)
print("COMPARISON: photos_metadatas_filtered.csv vs. photos_metadatas_with_hist_rgb128.parquet")
print("=" * 80)

# 1. SHAPE COMPARISON
print("\n1. SHAPE COMPARISON:")
print(f"   CSV (df):       {df.shape[0]:,} rows × {df.shape[1]} columns")
print(f"   Parquet:        {df_parquet.shape[0]:,} rows × {df_parquet.shape[1]} columns")
print(f"   Row difference: {df_parquet.shape[0] - df.shape[0]:,}")
print(f"   Col difference: {df_parquet.shape[1] - df.shape[1]} columns")

# 2. COLUMN COMPARISON
print("\n2. COLUMN ANALYSIS:")
csv_cols = set(df.columns)
parquet_cols = set(df_parquet.columns)

new_cols = parquet_cols - csv_cols
removed_cols = csv_cols - parquet_cols
common_cols = list(csv_cols & parquet_cols)

print(f"   Common columns: {len(common_cols)}")
print(f"   New columns in parquet: {len(new_cols)}")
if new_cols:
    print(f"      {sorted(new_cols)}")
print(f"   Removed columns from CSV: {len(removed_cols)}")
if removed_cols:
    print(f"      {sorted(removed_cols)[:10]}{'...' if len(removed_cols) > 10 else ''}")

# 3. DATA TYPE COMPARISON FOR COMMON COLUMNS
print("\n3. DATA TYPE CHANGES (for common columns):")
dtype_changes = []
for col in sorted(common_cols):
    df_dtype = df[col].dtype
    parquet_dtype = df_parquet[col].dtype
    if df_dtype != parquet_dtype:
        dtype_changes.append((col, str(df_dtype), str(parquet_dtype)))

if dtype_changes:
    print(f"   Found {len(dtype_changes)} type changes:")
    for col, old_type, new_type in dtype_changes[:10]:
        print(f"      {col}: {old_type} → {new_type}")
else:
    print("   No data type changes for common columns")

# 4. SAMPLE COMPARISON OF NEW COLUMNS
print("\n4. NEW COLUMNS IN PARQUET FILE:")
if new_cols:
    print(f"   Total new columns: {len(new_cols)}")
    print(f"   Sample of new columns:")
    for col in sorted(new_cols)[:15]:
        print(f"      - {col}")
        print(f"        Type: {df_parquet[col].dtype}, Non-null: {df_parquet[col].notna().sum():,}")
else:
    print("   No new columns")

# 5. COMPLETENESS COMPARISON FOR COMMON COLUMNS
print("\n5. DATA COMPLETENESS (for common columns):")
csv_completeness = (1 - df[common_cols].isnull().sum() / len(df)) * 100
parquet_completeness = (1 - df_parquet[common_cols].isnull().sum() / len(df_parquet)) * 100

completeness_diff = (parquet_completeness - csv_completeness).abs()
cols_with_changes = completeness_diff[completeness_diff > 0.1].sort_values(ascending=False)

if len(cols_with_changes) > 0:
    print(f"   Columns with completeness changes:")
    for col in cols_with_changes.head(10).index:
        print(f"      {col}: CSV={csv_completeness[col]:.2f}% → Parquet={parquet_completeness[col]:.2f}%")
else:
    print("   No significant completeness changes")

print("\n" + "=" * 80)
```
